# EDA for our Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
url_df = pd.read_csv('fraudish_urls.csv')
```

## Import our dataset of fraud related articles with a cosine simliarity of .60

```python
#Getting the five most common matched terms
common_terms = url_df['matched_terms'].value_counts().head(5)
print("Five most common matched terms:")
print(common_terms)
```

```python
#3d visualization of titles based on their cosine similarity scores
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
scat = ax.scatter(url_df.index, url_df['best_cosine'], np.random.rand(len(url_df)), c=url_df['best_cosine'], cmap='viridis')
ax.set_xlabel('Index')
ax.set_ylabel('Cosine Similarity')
ax.set_zlabel('Random Z')
fig.colorbar(scat, ax=ax, label='Cosine Similarity')
plt.title('3D Scatter Plot of Titles by Cosine Similarity')
plt.show()
```
